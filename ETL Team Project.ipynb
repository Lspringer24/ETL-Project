```python
# importing the engine

import pandas as pd
from sqlalchemy import create_engine
```

```python
# importing and reading the File

csv_file = "./data.csv"
ufc_data_df = pd.read_csv(csv_file)
ufc_data_df.head()
```

```python
# importing and reading the File

excel_file = "./ALL UFC FIGHTERS 2_23_2016 SHERDOG.xlsx"
ufc_fighters_df = pd.read_excel(excel_file)
ufc_fighters_df.head()
```

```python
# buliding the table from csv file


new_ufc_data_df = ufc_data_df[['B_Age', 'B_Location', 'B_Name', 'B_Weight', 'winby']].copy()
new_ufc_data_df.head()
```

```python
# buliding the table from Excel file

new_ufc_fighters_df = ufc_fighters_df[['name', 'height', 'class']].copy()
new_ufc_fighters_df.head()
```

```python
# Merging both tables


ufc_df = pd.merge(new_ufc_data_df, new_ufc_fighters_df,left_on="B_Name", right_on="name") 
ufc_df.head()
```

```python
# Droping the B_Name column


ufc_df.drop('B_Name' ,axis=1).head()
```

```python
# New data set1

new_ufc_df = ufc_df[['B_Age', 'B_Location', 'B_Weight', 'winby', 'name', 'height', 'class']].copy()
new_ufc_df.head()
```

```python
# droping duplicate rows

new_ufc_df.drop_duplicates(subset=['B_Age', 'B_Location', 'B_Weight', 'winby', 'name', 'height', 'class', 'winby', 'name'], keep='first').head()
```

```python
#New data set with duplicate rows droped

new_ufc_df1 = new_ufc_df.drop_duplicates(subset=['B_Age', 'B_Location', 'B_Weight', 'winby', 'name', 'height', 'class', 'winby', 'name'], keep='first')
new_ufc_df1.head()
```

```python
new_ufc_df2 = new_ufc_df1.winby.value_counts()
pd.DataFrame(new_ufc_df2)
```
